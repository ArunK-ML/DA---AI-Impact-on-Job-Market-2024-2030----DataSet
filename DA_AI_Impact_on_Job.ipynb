```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Create 3 subplots
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# 1. Salary vs Automation Risk
sns.scatterplot(
    data=df,
    x="Automation Risk (%)",
    y="Median Salary (USD)",
    hue="Industry",
    ax=axes[0]
)
axes[0].set_title("Salary vs Automation Risk", fontsize=14)

# 2. Growth Rate by AI Impact Level
sns.boxplot(
    data=df,
    x="AI Impact Level",
    y="Growth_Rate",
    palette="Set2",
    ax=axes[1]
)
axes[1].set_title("Growth Rate by AI Impact Level", fontsize=14)

# 3. Remote Work Trends by Job Title (Top 10 most frequent job titles)
top_jobs = df['Job Title'].value_counts().head(10).index
filtered_df = df[df['Job Title'].isin(top_jobs)]
sns.boxplot(
    data=filtered_df,
    x="Remote Work Ratio (%)",
    y="Job Title",
    ax=axes[2]
)
axes[2].set_title("Remote Work Trends by Job Title", fontsize=14)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Full Code

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/AI Impact on Job Market (2024–2030).csv")

# Fill missing values
for column in df.columns:
    if df[column].isnull().sum() > 0:
        if df[column].dtype == 'object':
            df[column].fillna(df[column].mode()[0], inplace=True)
        else:
            df[column].fillna(df[column].mean(), inplace=True)

# Set plot style
sns.set(style="whitegrid")

# Create 3 subplots
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# 1. Salary vs Automation Risk
sns.scatterplot(
    data=df,
    x="Automation Risk (%)",
    y="Median Salary (USD)",
    hue="Industry",
    ax=axes[0]
)
axes[0].set_title("Salary vs Automation Risk", fontsize=14)

# 2. Growth Rate by AI Impact Level
sns.boxplot(
    data=df,
    x="AI Impact Level",
    y="Growth_Rate",
    palette="Set2",
    ax=axes[1]
)
axes[1].set_title("Growth Rate by AI Impact Level", fontsize=14)

# 3. Remote Work Trends by Job Title (Top 10 most frequent job titles)
top_jobs = df['Job Title'].value_counts().head(10).index
filtered_df = df[df['Job Title'].isin(top_jobs)]
sns.boxplot(
    data=filtered_df,
    x="Remote Work Ratio (%)",
    y="Job Title",
    ax=axes[2]
)
axes[2].set_title("Remote Work Trends by Job Title", fontsize=14)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Salary vs Automation Risk

plt.figure(figsize=(10, 6))
plt.hexbin(df["Automation Risk (%)"], df["Median Salary (USD)"], gridsize=30, cmap="viridis")
plt.colorbar(label='Counts')
plt.xlabel("Automation Risk (%)")
plt.ylabel("Median Salary (USD)")
plt.title("Hexbin: Salary vs Automation Risk")
plt.show()
```

```python
sns.jointplot(
    data=df,
    x="Automation Risk (%)",
    y="Median Salary (USD)",
    kind="kde",
    fill=True,
    cmap="mako",
    height=8
)
plt.suptitle("KDE Plot: Salary vs Automation Risk", y=1.02)
plt.show()
```

```python
#Growth Rate by AI Impact Level

#Violin Plot:

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="AI Impact Level", y="Growth_Rate", palette="Set3")
plt.title("Violin Plot: Growth Rate by AI Impact Level")
plt.show()
```

```python
# Swarm Plot:

plt.figure(figsize=(10, 6))
sns.swarmplot(data=df, x="AI Impact Level", y="Growth_Rate", palette="Set2", size=2)
plt.title("Swarm Plot: Growth Rate by AI Impact Level")
plt.show()
```

```python
# Remote Work Trends by Job Title

# Horizontal Bar Plot (Mean Remote Work Ratio):

top_jobs = df['Job Title'].value_counts().head(10).index
filtered_df = df[df['Job Title'].isin(top_jobs)]

mean_remote = filtered_df.groupby('Job Title')["Remote Work Ratio (%)"].mean().sort_values()

plt.figure(figsize=(10, 6))
mean_remote.plot(kind='barh', color='skyblue')
plt.xlabel("Average Remote Work Ratio (%)")
plt.title("Bar Chart: Average Remote Work by Job Title")
plt.show()
```

```python
# Education vs Median Salary

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Required Education", y="Median Salary (USD)", estimator='mean', palette="coolwarm")
plt.title("Average Salary by Required Education")
plt.xticks(rotation=45)
plt.show()
```

```python
# Industry vs Projected Openings

top_industries = df.groupby("Industry")["Projected Openings (2030)"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_industries.plot(kind='barh', color='orange')
plt.title("Top 10 Industries by Projected Job Openings (2030)")
plt.xlabel("Projected Openings")
plt.show()
```

```python
# Industry vs Job Openings (2030)

# Group by Industry and sum Projected Openings

industry_openings = df.groupby("Industry")["Projected Openings (2030)"].sum().sort_values(ascending=True)

# Plot horizontal bar chart

plt.figure(figsize=(10, 6))
industry_openings.plot(kind='barh', color='steelblue')
plt.title("Projected Job Openings by Industry (2030)")
plt.xlabel("Projected Openings (2030)")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("AI Impact on Job Market (2024–2030).csv")  # Update path if needed

# Fill missing values
for column in df.columns:
    if df[column].isnull().sum() > 0:
        if df[column].dtype == 'object':
            df[column].fillna(df[column].mode()[0], inplace=True)
        else:
            df[column].fillna(df[column].mean(), inplace=True)

# Select only numeric (continuous) columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    square=True
)
plt.title("Correlation Matrix of Continuous Variables")
plt.tight_layout()
plt.show()
```
